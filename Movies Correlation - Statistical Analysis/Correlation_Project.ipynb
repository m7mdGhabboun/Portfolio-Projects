```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
df = pd.read_csv(r"C:\Users\Admin\OneDrive - University Of Jordan\movies.csv")
```

```python
df.head()
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print(col + ' - ' + str(pct_missing) + '%')
```

```python
df2 = df.dropna().copy()
```

```python
for col in df2.columns:
    pct_missing = np.mean(df2[col].isnull())
    print(col + ' - ' + str(pct_missing) + '%')
```

```python
df2.dtypes
```

```python
df2['votes'] = df2['votes'].astype('int64')

df2['budget'] = df2['budget'].astype('int64')

df2['gross'] = df2['gross'].astype('int64')
```

```python
df2.head()
```

```python
df2['yearcorrect'] = df2['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
df2.head()
```

```python
df3 = df2.drop_duplicates().copy()
```

```python
plt.scatter(x = df3['budget'], y = df3['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget For Film')
plt.ylabel('Gross Earnings')

plt.show()
```

```python
df3.sort_values(by = ['gross'], ascending = False).head()
```

```python
sns.regplot(x = 'budget', y = 'gross', data = df3, scatter_kws = {"color" : "red"}, line_kws = {"color" : "blue"})
plt.show()
```

```python
df3.corr(method = 'pearson', numeric_only = True) # pearson, kendall, spearman
```

```python
corr_matrix = df3.corr(method = 'pearson', numeric_only = True)
sns.heatmap(corr_matrix, annot = True)
plt.title('Correlation Matrix For Numerical Features')
plt.xlabel('Movie Feature')
plt.ylabel('Movie Feature')

plt.show()
```
