```python
import pandas as pd
import math
pd.set_option('display.max.row', 10)

kepler = pd.read_csv(r"C:\Users\Admin\OneDrive - University Of Jordan\kepler_copy.csv", index_col = False)
kepler
```

```python
# STAGE 1 : REMOVE COLUMNS

kepler1 = kepler.drop(columns = ['rowid', 'kepid', 'kepoi_name', 'kepler_name',
                      'koi_pdisposition', 'koi_score'])

kepler1
```

```python
# STAGE 2 : REMOVE FASLE POSITIVES
# CANCEL THIS STAGE

kepler2 = kepler1
```

```python
# STAGE 3 : CONFIRMED -> 0
#           CANDIDATE -> 1
#           FALSE POSITIVE -> 2

binary_incoding = {
    "CONFIRMED" : "0",
    "CANDIDATE" : "1",
    "FALSE POSITIVE" : "2"	
}
kepler3 = kepler2.replace(binary_incoding)

kepler3
```

```python
# STAGE 4 : REPLACE MISSING VALUES IN 'koi_tce_delivname' WITH THE MODE
most_common_value = kepler3['koi_tce_delivname'].mode()[0]
kepler3['koi_tce_delivname'] = kepler3['koi_tce_delivname'].fillna(most_common_value)

kepler4 = kepler3

kepler4
```

```python
# STAGE 5 : FILL NULL VALUES WITH THEIR MEAN FOR NUMERICAL COLUMNS, AND WITH MODE FOR CATEGORICAL COLUMNS :

# FOR CATEGORICAL:
for col in kepler4.columns:
    if kepler4[col].dtype == 'object' or kepler4[col].dtype.name == 'category':
        mode_value = kepler4[col].mode()[0]
        kepler4[col] = kepler4[col].fillna(mode_value)
        print("col replaced")


# FOR NUMERICAL :

for col in kepler4.columns:
    if kepler4[col].dtype in ['float64', 'int64']:
        mean_value = kepler4[col].mean()
        if math.isnan(mean_value):
            kepler4.drop(col,axis = 1, inplace = True)
            print("col dropped")

    

kepler5 = kepler4

kepler5
```

```python
# STAGE 6 : DROP USELESS COLUMNS
validation_features = [
    "koi_disposition",
    "koi_score",
    "koi_fpflag_nt", "koi_fpflag_ss", "koi_fpflag_co", "koi_fpflag_ec",
    "koi_max_sngle_ev", "koi_max_mult_ev",
    "koi_model_snr",
    "koi_ldm_coeff1", "koi_ldm_coeff2", "koi_ldm_coeff3", "koi_ldm_coeff4",
    "koi_quarters",
    "koi_bin_oedp_sig",
    "koi_steff", "koi_slogg", "koi_smet", "koi_srad", "koi_smass",
    "ra", "dec",
    "koi_kepmag", "koi_gmag", "koi_rmag", "koi_imag", "koi_zmag",
    "koi_jmag", "koi_hmag", "koi_kmag"
]
detection_features = [
    "koi_period",    
    "koi_time0bk",
    "koi_eccen",
    "koi_impact",
    "koi_duration",
    "koi_depth",
    "koi_ror",
    "koi_srho",
    "koi_prad",
    "koi_sma",
    "koi_incl",
    "koi_teq",
    "koi_insol",
    "koi_dor"
]
# wanted_columns = [
#     'koi_disposition', 'koi_period','koi_time0bk','koi_eccen','koi_impact',
#     'koi_duration','koi_depth','koi_ror','koi_srho','koi_prad',
#     'koi_sma','koi_incl','koi_teq','koi_insol','koi_dor'
# ]
wanted_columns = validation_features + detection_features
for col in kepler5.columns:
    if col not in wanted_columns:
        kepler5.drop(col, axis= 1, inplace = True)
        print("col dropped")
```
